```python
#Import libraries for data handling
import pandas as pd
```

```python
#Import libraries for data clustering
from sklearn.cluster import KMeans
```

```python
#Import libraries for data visualization
import matplotlib.pyplot as plt
```

```python
#For loading the dataset from the CSV file
df=pd.read_csv('/content/6d_dataset_no_clusters.csv')
```

```python
df.head()
```

```python
# used to display any visualizations or plots that you create using the matplotlib library
%matplotlib inline
```

```python
#creates a list called numClusters and assigns it the values 1, 2, 3, 4, 5, and 6.
numClusters=[1,2,3,4,5,6]
```

```python
 #creates an empty list called SSE.
SSE=[]
```

```python
#Import libraries for data clustering
from sklearn.cluster import KMeans
```

```python
for k in numClusters:
  #creates a KMeans object and specifies the number of clusters to be k.
  k_means=KMeans(n_clusters=k)
  #fits the KMeans model to the data contained in the dataframe df.
  k_means.fit(df)
  #calculates the sum of squared errors (SSE) for the current KMeans model and appends it to the list SSE.
  SSE.append(k_means.inertia_)
```

```python
#Displays the list SSE
SSE
```

```python
  #Plots the data
  plt.plot(numClusters,SSE)
  plt.xlabel('Number of Clusters')
  plt.ylabel('SSE')
```

```python
#Extract the features from the dataset
X = df.values
```

```python
#Used to set the desired number of clusters
num_clusters = 3
```

```python
#Used to perform k-means clustering for initializing and fitting KMeans model
kmeans = KMeans(n_clusters=num_clusters)
kmeans.fit(X)
```

```python
#Get cluster labels and centroids
labels = kmeans.labels_
#Assigns each data point to a cluster
centroids = kmeans.cluster_centers_
```

```python
#Add a new column to the dataframe to store the assigned cluster labels
df['Cluster']= labels
```

```python
#Create a scatterplot to visualize the cluster and centroids
plt.scatter(X[:,0], X[:, 1],
            c=labels, cmap='viridis',
            edgecolors='k', marker='o', s=50)
plt.scatter(centroids[:, 0],
            centroids[:, 1], c='red',
            marker='X', s=200,
            label='Centroids')
plt.title('K_Means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```
